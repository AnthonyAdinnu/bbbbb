```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
%matplotlib inline
```

```python
df = pd.read_csv("mtcars.csv")
```

```python
df.head()
```

```python
df.info()
```

```python
df.shape
```

```python
#barplot
res = sns.barplot(x="cyl", y="mpg", data=df,palette = "tab10")

# Show the plot
plt.show()
```

```python
#countplot
# Vertical countplots
sns.countplot(x = 'cyl', data = df, palette ="Set1" )
```

```python
#Horizontal countplots
sns.countplot(y = "gear", data = df, palette ="rocket")
```

```python
#group countplotl
#'hue' is used to visualize the data of different categories in one plot. 'palette' is used to change the colour of the plot.
# arr = arr.astype(int) 
# sns.countplot(x = 'gear', hue ='cyl', data = df ,palette ="Set2")
# 
```

```python
#distributionplot is use for continues data
sns.displot(df.mpg, bins = 5, color = 'g')
```

```python
df2 = df.drop(df.columns[0], axis=1)
```

```python
df2.head()
```

```python
# heatmap in Seaborn is a data visualization tool used to represent the magnitude of values in a matrix or 2D dataset using color.
#It is particularly useful for identifying patterns, correlations, and distributions within the data.
sns.heatmap(df2.corr(), cbar = True, linewidths =0.5)
```
